### divide, floor_divide
* np.devide(x1, x2)
    * x1: 피제수(Dividend )
    * x2: 제수(Divisor)
* a÷b=c
    * a: 피제수
    * b: 제수
    * c: 몫

```python
# In Python 3.0, // is the floor division operator and / the true division operator
```

```python
x = np.arange(5)
display(x)
np.divide(x, 4)
```

```python
display(x/4) # 일반적 나눗셈의 결과
# python2.X에서는 array([0, 0, 0, 0, 1]) 결과
x//4 # floor division operator
```

### power
    * 첫 번째 원소를 두 번째 원소만큼 제곱한다.
    * x1 and x2 must be broadcastable to the same shape
    * an integer type raised to a negative integer power will raise a ValueError.

```python
x1 = range(6)
np.power(x1, 3)
```

```python
np.power(x1, -3)
```

```python
x2 = [1.0, 2.0, 3.0, 3.0, 2.0, 1.0]
np.power(x1, x2)
```

```python
x3 = np.array([[1, 2, 3, 3, 2, 1], [1, 2, 3, 3, 2, 1]])
np.power(x1, x3)
```

```python
x3=x3.reshape(2,2,3)
np.power(x1, x3)
#page 128
```

### maximum, fmax
* 두 원소 중 큰 값을 반환
* where을 사용하면, np.where(x1 >= x2, x1, x2) 와 결과가 동일하다

```python
np.maximum([2, 3, 4], [1, 5, 2])
```

```python
x1=np.array([2, 3, 4])
x2=np.array([1, 5, 2])
np.where(x1 >= x2, x1, x2)
```

```python
np.eye(2)
```

```python
np.maximum(np.eye(2), [0.5, 2])
```

```python
np.maximum([np.NAN, 0, np.NAN], [0, np.NAN, np.NAN])
```

```python
np.maximum(np.Inf, 1)
```

```python
np.fmax([2, 3, 4], [1, 5, 2])
```

```python
np.fmax(np.eye(2), [0.5, 2])
```

```python
np.fmax([np.nan, 0, np.nan],[0, np.nan, np.nan])
#nan이 무시된다
```

### copysign
* 첫 번재 배열 원소 기호를 두 번째 배열 원소 기호로 바꾼다.
* 같은 길이면 결과값도 같은 길이
* x2가 스칼라라면 x1의 sign이 모두 동일하게 바뀐다

```python
np.copysign(1.3, -1)
```

```python
np.copysign([-1, 0, 1], -3)
```

```python
 np.copysign([20, 30, 0], np.arange(3)-1)
```

### 배열집합 연산

* 가장 자주 사용되는 함수: np.unique
    * 중복 제거

```python
a = np.array([1, 2, 6, 4, 2, 3, 2])
u, indices = np.unique(a, return_inverse=True)
```

```python
display(u)
# 원래 원소의 인덱스를 알려준다.
indices
```

* intersect1d: 공통 존재 원소를 정렬 후 반환

```python
np.intersect1d([1, 3, 4, 3, 5], [3, 1, 2, 1, 5, 5])
```

* union1d: 합집합을 반환한다.

```python
np.union1d([-1, 0, 1], [-2, 0, 2])
```

```python
np.union1d([1, 3, 4, 3],[6, 3, 4, 2])
```

* in1d: x의 원소 중 y의 원소를 포함하는지 나타나는 불리언 배열 반환

```python
test = np.array([0, 1, 2, 5, 0])
states = [0, 2]
```

```python
mask = np.in1d(test, states)
```

```python
mask
```

```python
a = np.array([1, 2, 3, 2, 4, 1])
b = np.array([3, 4, 5, 6])
np.setdiff1d(a, b)
# 차집합을 반환한다.
# b에 속하지 않은 원소를 a의 1차원 배열로 정렬한다. 
```

```python
a = np.array([1, 2, 3, 2, 4])
b = np.array([2, 3, 5, 7, 5])
np.setxor1d(a,b)
# 한 배열에는 포함되지만, 두 배열 모두에는 포함되지 않는 원소들의 집합
```
