```python
import numpy as np
import matplotlib.pyplot as plt
score=np.loadtxt('python_score.csv', delimiter=',', dtype='float64')
# 정답
answer1=score[:,10:]
```

```python
rset1=score[:, :10]
b1=np.ones(22, dtype='uint8')
b1=b1*4
b1[20:]=1
bool_answer1=(answer1==rset1)
dd1=bool_answer1*1
dd1[20:]=rset1[20:]
result=np.dot(dd1.T, b1)
result
```

```python
plt.plot(result)
```

```python
plt.yticks(np.arange(0, 100, step=10))
plt.bar(np.arange(10) ,result)
```
